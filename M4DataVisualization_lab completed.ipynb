# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
QUERY = """
SELECT Respondent, ConvertedComp
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df_ConvertedComp = pd.read_sql_query(QUERY,conn)

df_ConvertedComp['ConvertedComp'].plot(kind='hist', figsize=(8, 5))
plt.title('Histogram of Converted Compensation')
plt.ylabel('Number of Respondents') 
plt.xlabel('Annual $USD') 

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Respondent, Age
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df_Age = pd.read_sql_query(QUERY,conn)

df_Age['Age'].plot(kind='box', figsize=(8, 6))

plt.title('Box plot of Age')
plt.ylabel('Age')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Respondent, Age, WorkWeekHrs
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df_AgeWWH = pd.read_sql_query(QUERY,conn)

df_AgeWWH[['Age', 'WorkWeekHrs']].plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10, 6), color='darkblue')

plt.title('Scatter Plot of Age and WorkWeekHrs')
plt.xlabel('Age in Years')
plt.ylabel('Work Week Hours')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df_bubble = pd.read_sql_query(QUERY,conn)

norm_age = (df_bubble['Age'] - df_bubble['Age'].min())/(df_bubble['Age'].max() - df_bubble['Age'].min())

df_bubble[['CodeRevHrs', 'WorkWeekHrs']].plot(kind='scatter', y='CodeRevHrs', x='WorkWeekHrs', figsize=(10, 6), alpha = 0.5, color='darkblue', s=norm_age * 1000 + 10)

plt.title('Bubble Plot')
plt.ylabel('CodeRevHrs')
plt.xlabel('Work Week Hours')
plt.legend(['Age'], loc='upper right', fontsize='x-large')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

df_db = pd.read_sql_query(QUERY,conn)
df_db.head(10)
```

```python
QUERY = """
SELECT *, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

df_db = pd.read_sql_query(QUERY,conn)
df_db.set_index('DatabaseDesireNextYear', inplace=True) 
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
df_db['count'].plot(kind='pie',
                            figsize=(10, 6),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90, labels=None, pctdistance=1.12, colors=colors_list,     
                            )

# scale the title up by 12% to match pctdistance
plt.title('Top 5 databases that respondents wish to learn', y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=df_db.index, loc='upper right') 

plt.show()
```

In the list of most popular languages respondents wish to learn next year, what is the rank of Python?

```python
QUERY = """
SELECT *, COUNT(*) as count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

df_lang = pd.read_sql_query(QUERY,conn)
df_lang.set_index('LanguageDesireNextYear', inplace=True) 
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
df_lang['count'].plot(kind='pie',
                            figsize=(10, 6),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90, labels=None, pctdistance=1.12, colors=colors_list,     
                            )

# scale the title up by 12% to match pctdistance
plt.title('Top 5 Language that respondents wish to learn', y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=df_lang.index, loc='upper right') 

plt.show()
```

How many respondents indicated that they currently work with 'SQL'?

How many respondents indicated that they work on 'MySQL' only?

```python
QUERY = """
SELECT * 
FROM DatabaseWorkedWith
ORDER BY Respondent
"""
df = pd.read_sql_query(QUERY,conn)

counts = df.groupby('Respondent').count() #counts: number databases each respondent has worked with

for i in range(len(counts)):
    if counts.values[i] == 1:
        uniques.append(counts.index[i]) #uniques: respondents who have worked with one database

uniq_mysql = 0

for j in range(len(df)):
    if df['Respondent'][j] in uniques:
        if df['DatabaseWorkedWith'][j] == 'MySQL':
            uniq_mysql = uniq_mysql + 1 #uniq_mysql: respondents who have worked with MySQL only

uniq_mysql    
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY,conn)
df1 = df.groupby('Age').median()

df1.plot(kind='bar', figsize=(10, 6), stacked=True)

plt.title('Stacked Bar Chart of Median WorkWeekHrs and CodeRevHrs for Those Age 30 to 35')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age FROM master
WHERE Age BETWEEN 45 AND 60
"""
df_line = pd.read_sql_query(QUERY,conn)
df_line.dropna(inplace=True)
df_line = df_line.groupby('Age',axis=0).median()

df_line.plot(kind='line')

plt.title('Average Compensation Based on Age Groups')
plt.ylabel('Converted Compensation (Annual USD)')
plt.xlabel('Age')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
SELECT MainBranch, COUNT(*) AS Frequency
FROM master
GROUP BY MainBranch

"""
df_bar = pd.read_sql_query(QUERY,conn)

df_bar.plot(kind='barh')
plt.show()
```

Majority of the survey responders are
